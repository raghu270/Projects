# C) Model Building:

```python
a={"Yes":1,"No":0}
data["Churn"]=data["Churn"].map(a)
x = data["tenure"]
y = data["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

# Model - 1

```python
model_1 = Sequential()
model_1.add(Dense(12,activation='relu',input_shape=(1,)))
model_1.add(Dense(8,activation='relu'))
model_1.add(Dense(1,activation='sigmoid'))
```

```python
model_1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model_1.fit(x_train,y_train,epochs=150,validation_data=(x_test,y_test))
```

```python
y_pred_probs = model_1.predict(x_test)
```

```python
y_pred = (y_pred_probs > 0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
plt.plot(history.history['val_accuracy']) 
plt.plot(history.history['accuracy'])
plt.title("Accuracy v/s Epochs")
plt.show()
```

# Model -2

```python
x = data["tenure"]
y = data["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=47)
```

```python
from keras.layers import Dropout
model_2 = Sequential() 
model_2.add(Dense(12, input_dim=1, activation='relu')) 
model_2.add(Dropout(0.3)) 
model_2.add(Dense(8, activation='relu')) 
model_2.add(Dropout(0.2)) 
model_2.add(Dense(1, activation='sigmoid')) 
```

```python
model_2.summary()
```

```python
model_2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
history = model_2.fit(x_train, y_train, epochs=150,validation_data=(x_test,y_test))
```

```python
y_pred_probs = model_2.predict(x_test)
y_pred = (y_pred_probs > 0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python

plt.plot(history.history['accuracy'])
plt.title("Accuracy v/s Epochs")
plt.show()
```

# Model -3

```python
x=data[['MonthlyCharges','tenure','TotalCharges']]#Features 
y=data[['Churn']]#Target 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42) 
model_3 = Sequential() 
model_3.add(Dense(12, input_dim=3, activation='relu')) 
model_3.add(Dense(8, activation='relu')) 
model_3.add(Dense(1, activation='sigmoid')) 
model_3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
history = model_3.fit(x_train, y_train, epochs=150,validation_data=(x_test,y_test)) 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
